```python
# Word Embeddings Visualization
# This notebook demonstrates how word embeddings work by visualizing words in a 2D space

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import gensim.downloader as api
import os
from matplotlib.pyplot import figure

# Set larger figure size for better visualization
plt.rcParams["figure.figsize"] = (12, 10)

# Download the GloVe Twitter 200d embeddings
# This may take a moment to download if you don't have it already
print("Loading the GloVe Twitter 200d embeddings...")
glove_twitter = api.load("glove-twitter-200")
print("Embeddings loaded!")

# List of words to visualize
words = ['king', 'queen', 'man', 'woman', 'child', 'throne', 'crown', 'royalty']

# Check if all words are in the vocabulary
unavailable_words = [word for word in words if word not in glove_twitter.key_to_index]
if unavailable_words:
    print(f"Warning: These words are not in the vocabulary: {unavailable_words}")
    words = [word for word in words if word in glove_twitter.key_to_index]

# If we have no valid words, we can't proceed
if not words:
    raise ValueError("None of the requested words are in the vocabulary!")

# Extract word vectors for our words
word_vectors = [glove_twitter[word] for word in words]

# Convert list of vectors to a 2D numpy array
X = np.array(word_vectors)

# Apply PCA to reduce dimensions to 2D
pca = PCA(n_components=2)
result = pca.fit_transform(X)

# Create a scatterplot of the words
plt.figure(figsize=(12, 8))
plt.scatter(result[:, 0], result[:, 1], c='steelblue', edgecolors='k', s=100)

# Add word labels
for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]), xytext=(5, 2),
                 textcoords='offset points', ha='right', va='bottom',
                 fontsize=14)

# Add title and labels
plt.title('2D Representation of Word Embeddings', fontsize=18)
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

# Add percentage of variance explained
explained_variance = pca.explained_variance_ratio_
plt.figtext(0.5, 0.01, f'Explained variance: PC1 = {explained_variance[0]:.2%}, PC2 = {explained_variance[1]:.2%}',
            ha='center', fontsize=12)

# Make the plot prettier
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Let's also visualize the relationships between words with arrows
plt.figure(figsize=(12, 8))
plt.scatter(result[:, 0], result[:, 1], c='steelblue', edgecolors='k', s=100)

# Add word labels again
for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]), xytext=(5, 2),
                 textcoords='offset points', ha='right', va='bottom',
                 fontsize=14)

# Draw arrows between related pairs to show relationships
if all(w in words for w in ['king', 'queen', 'man', 'woman']):
    # Indices of the words we want to connect
    king_idx = words.index('king')
    queen_idx = words.index('queen')
    man_idx = words.index('man')
    woman_idx = words.index('woman')

    # Draw arrow from man to woman
    plt.arrow(result[man_idx, 0], result[man_idx, 1],
              result[woman_idx, 0] - result[man_idx, 0],
              result[woman_idx, 1] - result[man_idx, 1],
              head_width=0.01, head_length=0.01, fc='red', ec='red',
              length_includes_head=True, alpha=0.7)

    # Draw arrow from king to queen
    plt.arrow(result[king_idx, 0], result[king_idx, 1],
              result[queen_idx, 0] - result[king_idx, 0],
              result[queen_idx, 1] - result[king_idx, 1],
              head_width=0.01, head_length=0.01, fc='green', ec='green',
              length_includes_head=True, alpha=0.7)

    # Add a legend
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color='red', lw=2, label='man ‚Üí woman'),
        Line2D([0], [0], color='green', lw=2, label='king ‚Üí queen')
    ]
    plt.legend(handles=legend_elements, loc='lower right')

plt.title('Word Embedding Relationships', fontsize=18)
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Extra: Let's calculate and print some vector arithmetic to demonstrate embedding properties
print("\nWord Vector Arithmetic Examples:")

if all(w in glove_twitter.key_to_index for w in ['king', 'man', 'woman']):
    # king - man + woman should be close to queen
    result = glove_twitter.most_similar(
        positive=['king', 'woman'],
        negative=['man'],
        topn=3
    )
    print(f"\nking - man + woman ‚âà {result}")

if all(w in glove_twitter.key_to_index for w in ['throne', 'king']):
    # Find words similar to throne
    similar_to_throne = glove_twitter.most_similar('throne', topn=5)
    print(f"\nWords similar to 'throne': {similar_to_throne}")

    # Find words similar to king
    similar_to_king = glove_twitter.most_similar('king', topn=5)
    print(f"\nWords similar to 'king': {similar_to_king}")

# Calculate and display the cosine similarities between all pairs of words
print("\nCosine Similarities between words:")
for i, word1 in enumerate(words):
    for word2 in words[i+1:]:
        similarity = glove_twitter.similarity(word1, word2)
        print(f"{word1} ‚Üî {word2}: {similarity:.4f}")
```
